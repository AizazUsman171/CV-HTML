```python
mean_Age = df['Age'].mean() 
df['Age'].fillna(mean_Age,inplace=True)
# print(df['Age']) //Here i used mean because its a no and this method is most commonly used in stats
df['Age'].isnull()
```

```python
df['Embarked'].fillna(method='ffill', inplace=True)
df['Embarked'].isnull() #Here bfill is used because the data is sequential
```

```python
df['Cabin'].fillna(method='ffill', inplace=True)
df['Cabin'].isnull() #Here bfill is used because the data is sequential
```

```python
df.isnull()
```

```python
c=df['Sex'].value_counts()
print(c)
```

```python
age_groups = ['Children', 'Adult', 'Senior_Citizen']
age_ranges = [0, 18, 60, float('inf')]
df['Age_Group'] = pd.cut(df['Age'], bins=age_ranges, labels=age_groups)
df.head()
```

```python
df['Total_Family_Size'] = df['SibSp'] + df['Parch']
print(df)
```

```python
df['Is_Alone'] = df['Total_Family_Size'].apply(lambda x: 1 if x == 0 else 0)
print(df)
```
